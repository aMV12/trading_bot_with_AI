```python
import matplotlib.pyplot as plt # Подключаем matplotlib
import numpy as np # Подключаем numpy
import seaborn as sns
import csv
import requests
import time
import pandas as pd
from tensorflow import keras
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings("ignore")
# Рисовать графики сразу же
import time
from datetime import datetime
from threading import Thread
#Загружаем библиотеки
import pandas as pd #Пандас
import matplotlib.pyplot as plt #Отрисовка графиков
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler # проверить все
import numpy as np #Numpy
from tensorflow.keras.optimizers import Adam #Оптимизатор
from tensorflow.keras.models import Sequential, Model #Два варианты моделей
from tensorflow.keras.layers import concatenate, Input, Dense, Dropout, BatchNormalization, Flatten, Conv1D, Conv2D, LSTM, GlobalMaxPooling1D, MaxPooling1D, RepeatVector #Стандартные слои
from sklearn.preprocessing import StandardScaler, MinMaxScaler #Нормировщики
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator # для генерации выборки временных рядов
import ssl
import time
import json
import urllib
import hmac, hashlib
import requests
import statistics
#import discord
#from discord.ext import commands
import asyncio
import concurrent.futures
from urllib.parse import urlparse, urlencode
from urllib.request import Request, urlopen
from datetime import timezone
import time
from google.colab import drive
drive.mount('/content/drive')

class Binance():
    methods = {
            # public methods
            'ping':             {'url':'api/v1/ping', 'method': 'GET', 'private': False},
            'time':             {'url':'api/v1/time', 'method': 'GET', 'private': False},
            'exchangeInfo':     {'url':'api/v1/exchangeInfo', 'method': 'GET', 'private': False},
            'depth':            {'url': 'api/v1/depth', 'method': 'GET', 'private': False},
            'trades':           {'url': 'api/v1/trades', 'method': 'GET', 'private': False},
            'historicalTrades': {'url': 'api/v1/historicalTrades', 'method': 'GET', 'private': False},
            'aggTrades':        {'url': 'api/v1/aggTrades', 'method': 'GET', 'private': False},
            'klines':           {'url': 'api/v1/klines', 'method': 'GET', 'private': False},
            'ticker24hr':       {'url': 'api/v1/ticker/24hr', 'method': 'GET', 'private': False},
            'tickerPrice':      {'url': 'api/v3/ticker/price', 'method': 'GET', 'private': False},
            'tickerBookTicker': {'url': 'api/v3/ticker/bookTicker', 'method': 'GET', 'private': False},
            # private methods
            'createOrder':      {'url': 'api/v3/order', 'method': 'POST', 'private': True},
            'testOrder':        {'url': 'api/v3/order/test', 'method': 'POST', 'private': True},
            'orderInfo':        {'url': 'api/v3/order', 'method': 'GET', 'private': True},
            'cancelOrder':      {'url': 'api/v3/order', 'method': 'DELETE', 'private': True},
            'openOrders':       {'url': 'api/v3/openOrders', 'method': 'GET', 'private': True},
            'allOrders':        {'url': 'api/v3/allOrders', 'method': 'GET', 'private': True},
            'account':          {'url': 'api/v3/account', 'method': 'GET', 'private': True},
            'myTrades':         {'url': 'api/v3/myTrades', 'method': 'GET', 'private': True},
            # wapi
            'depositAddress':   {'url': 'wapi/v3/depositAddress.html', 'method':'GET', 'private':True},
            'withdraw':   {'url': 'wapi/v3/withdraw.html', 'method':'POST', 'private':True},
            'depositHistory': {'url': 'wapi/v3/depositHistory.html', 'method':'GET', 'private':True},
            'withdrawHistory': {'url': 'wapi/v3/withdrawHistory.html', 'method':'GET', 'private':True},
            'assetDetail': {'url': 'wapi/v3/assetDetail.html', 'method':'GET', 'private':True},
            'tradeFee': {'url': 'wapi/v3/tradeFee.html', 'method':'GET', 'private':True},
            'accountStatus': {'url': 'wapi/v3/accountStatus.html', 'method':'GET', 'private':True},
            'systemStatus': {'url': 'wapi/v3/systemStatus.html', 'method':'GET', 'private':True},
            'assetDust': {'url': 'sapi/v1/asset/dust', 'method':'POST', 'private':True},
            'dustLog': {'url': 'wapi/v3/userAssetDribbletLog.html', 'method':'GET', 'private':True},
            'assetAssetDividend': {'url': 'sapi/v1/asset/assetDividend', 'method':'GET', 'private':True},
            #sapi
            'marginTransfer': {'url': 'sapi/v1/margin/transfer', 'method': 'POST', 'private':True},
            'marginLoan': {'url': 'sapi/v1/margin/loan', 'method': 'POST', 'private': True},
            'marginLoanGet': {'url': 'sapi/v1/margin/loan', 'method': 'GET', 'private': True},
            'marginRepay': {'url': 'sapi/v1/margin/repay', 'method': 'POST', 'private': True},
            'marginRepayGet': {'url': 'sapi/v1/margin/repay', 'method': 'GET', 'private': True},
            'marginCreateOrder': {'url': 'sapi/v1/margin/order', 'method': 'POST', 'private':True},
            'marginCancelOrder': {'url': 'sapi/v1/margin/order', 'method': 'DELETE', 'private':True},
            'marginOrderInfo': {'url': 'sapi/v1/margin/order', 'method': 'GET', 'private':True},
            'marginAccount': {'url': 'sapi/v1/margin/account', 'method': 'POST', 'private':True},
            'marginOpenOrders': {'url': 'sapi/v1/margin/openOrders', 'method': 'GET', 'private':True},
            'marginAllOrders': {'url': 'sapi/v1/margin/allOrders', 'method': 'GET', 'private':True},
            'marginAsset': {'url': 'sapi/v1/margin/asset', 'method': 'POST', 'private':True},
            'marginPair': {'url': 'sapi/v1/margin/pair', 'method': 'POST', 'private':True},
            'marginPriceIndex': {'url': 'sapi/v1/margin/priceIndex', 'method': 'POST', 'private':True},
            'marginMyTrades': {'url': 'sapi/v1/margin/myTrades', 'method': 'GET', 'private':True},
            'marginMaxBorrowable': {'url': 'sapi/v1/margin/maxBorrowable', 'method': 'GET', 'private':True},
            'marginmaxTransferable': {'url': 'sapi/v1/margin/maxTransferable', 'method': 'GET', 'private':True},
            #futures
            'futuresExchangeInfo': {'url': 'fapi/v1/exchangeInfo', 'method': 'GET', 'private': False, 'futures': True},
            'futuresKlines': {'url': 'fapi/v1/klines', 'method': 'GET', 'private': False, 'futures': True},
            'futuresCreateOrder':      {'url': 'fapi/v1/order', 'method': 'POST', 'private': True, 'futures': True},
            'futuresAccount':      {'url': 'fapi/v1/account', 'method': 'POST', 'private': True, 'futures': True},
            'futuresBalance':      {'url': 'fapi/v1/balance', 'method': 'GET', 'private': True, 'futures': True},
            'futuresSymbolPriceTicker': {'url': 'fapi/v1/ticker/price', 'method': 'GET', 'private': True, 'futures': True},
            'futuresOrderInfo': {'url': 'fapi/v1/order', 'method': 'GET', 'private': True, 'futures': True},
            'futuresCancelOrder':      {'url': 'fapi/v1/order', 'method': 'DELETE', 'private': True, 'futures': True},
   }

    def __init__(self, API_KEY, API_SECRET):
        self.API_KEY = API_KEY
        self.API_SECRET = bytearray(API_SECRET, encoding='utf-8')
        self.shift_seconds = 0

    def __getattr__(self, name):
        def wrapper(*args, **kwargs):
            kwargs.update(command=name)
            return self.call_api(**kwargs)
        return wrapper

    def set_shift_seconds(self, seconds):
        self.shift_seconds = seconds

    def call_api(self, **kwargs):

        command = kwargs.pop('command')

        base_url ='https://api.binance.com/'
        if self.methods[command].get('futures'):
            base_url = 'https://fapi.binance.com/'
        api_url = base_url  + self.methods[command]['url']

        payload = kwargs
        headers = {}

        payload_str = urllib.parse.urlencode(payload)
        if self.methods[command]['private']:
            payload.update({'timestamp': int(time.time() + self.shift_seconds - 1) * 1000})
            payload_str = urllib.parse.urlencode(payload).encode('utf-8')
            sign = hmac.new(
                key=self.API_SECRET,
                msg=payload_str,
                digestmod=hashlib.sha256
            ).hexdigest()

            payload_str = payload_str.decode("utf-8") + "&signature="+str(sign)
            headers = {"X-MBX-APIKEY": self.API_KEY, "Content-Type":"application/x-www-form-urlencoded"}

        if self.methods[command]['method'] == 'GET' or self.methods[command]['url'].startswith('sapi'):
            api_url += '?' + payload_str

        response = requests.request(method=self.methods[command]['method'], url=api_url, data="" if self.methods[command]['method'] == 'GET' else payload_str, headers=headers)

        if 'code' in response.text:
            raise Exception(response.text)

        return response.json()

step_in_the_last = 1   # По скольки шагам в прошлое определяем тренд
n_steps = 100            # Серию из скольки дней анализируем
forvard_lag = 1         # Предсказание какого бара в будущем делаем
analyze_period = 100000  # Сколько баров в прошлое берем для тренировки модели (чтобы ихбежать тренировки на другой динамике рынка)
n_train = 1000         # Кол-во тренировочных данных
fix = 7                 # Кол-во деняг
last = True
transactions = 50
Commission_buy = 0.0002
Commission_sell = 0.0004
Old_Prediction = None
first = False
BUY_A = False
BUY_B = False
SELL_A = False
SELL_B = False
_API_KEY_='EvnNm9TKqsC2p48HsCdApx7c0UHgFLXByOL6hPjWIKeLS8TvOifSvZnqEa6mniWI'
_API_SECRET_='r1j02rVBir11OKOVeiPe2LvJ8klDC6bm3NSjNfNJSMKkwiTHc6iix2om4fbsIaOf'
_SYMBOL_ = 'ETHUSDT'
#BTCUSD ETHUSD LTCUSD
interval = 1
_INTERVAL_ = str(interval) + 'm'
_analyze_period_test_ = 555

bot = Binance(
    API_KEY=_API_KEY_,
    API_SECRET=_API_SECRET_
)

fixed_btc = fix / float(bot.tickerPrice(symbol=_SYMBOL_)['price'])
fixed_btc = 0.002
model = keras.models.load_model('/content/drive/My Drive/ETHUSD_best.h5')
#model = Variations[len(Variations)-1]["Model"]

def TimeStampNow():
  return time.time()

def TimeNow():
  now = datetime.now()
  return now.strftime('%Y-%m-%d %H:%M:%S')

def TimestampToNormal(inputTime):
  return datetime.utcfromtimestamp(inputTime/1000).strftime('%Y-%m-%d %H:%M:%S')

def NormalToTimestamp(inputTime):
  return int(datetime.strptime(inputTime, '%Y-%m-%d %H:%M:%S').timestamp()*1000)

def TimeSpent(msg,LocalTimeSpent_start):
  _TimeNow_ = TimeStampNow()
  LocalTimeSpent_end = _TimeNow_
  print(f'{msg}{round(LocalTimeSpent_end - LocalTimeSpent_start,6)} секунд')
  LocalTimeSpent_start = _TimeNow_
  return LocalTimeSpent_start

def MinetsTimestamp(minets):
  return minets*60*1000

def DataLoad(bot=bot):
  _TimeNow_ = TimeNow()
  _TimestampNow_ = TimeStampNow()
  TotalTimeSpent_start = _TimestampNow_
  LocalTimeSpent_start = _TimestampNow_
  TimestampNow = NormalToTimestamp(_TimeNow_)
  _MINETS_ = interval*(step_in_the_last+n_steps+forvard_lag+2)
  klines = bot.klines(
      symbol=_SYMBOL_,
      interval=_INTERVAL_,
      limit=_MINETS_,
      startTime=TimestampNow - MinetsTimestamp(_MINETS_),
      endTime=TimestampNow,
  )

  LocalTimeSpent_start = TimeSpent('Затраченное время на подгрузку бд из api: ',LocalTimeSpent_start)

  Close = []
  for Number in range(len(klines)):
    Close.append(float(klines[Number][4])) # На 4 позиции находится Close свечки

  LastKlines = klines[len(klines)-1][6]
  if LastKlines > NormalToTimestamp(TimeNow()):
    Close.pop(len(Close)-1)

  LocalTimeSpent_start = TimeSpent('Затраченное время на запись всех Close в list: ',LocalTimeSpent_start)

  data = {'Close': Close}
  data = pd.DataFrame(data)
  data = data.append({'Close':69,},ignore_index=True)

  LocalTimeSpent_start = TimeSpent('Затраченное время на преоброзование из list в DataFrame: ',LocalTimeSpent_start)
  TotalTimeSpent_end = TimeSpent('Полное затраченное время на загрузку и формирование бд: ',TotalTimeSpent_start)

  return TotalTimeSpent_start,data

def DataLoad_test(bot=bot):
  _TimeNow_ = TimeNow()
  _TimestampNow_ = TimeStampNow()
  TotalTimeSpent_start = _TimestampNow_
  LocalTimeSpent_start = _TimestampNow_
  TimestampNow = NormalToTimestamp(_TimeNow_)
  _MINETS_ = interval*(step_in_the_last+n_steps+forvard_lag+2+_analyze_period_test_)
  klines = bot.klines(
      symbol=_SYMBOL_,
      interval=_INTERVAL_,
      limit=_MINETS_,
      startTime=TimestampNow - MinetsTimestamp(_MINETS_),
      endTime=TimestampNow,
  )

  LocalTimeSpent_start = TimeSpent('Затраченное время на подгрузку бд из api: ',LocalTimeSpent_start)

  Close = []
  for Number in range(len(klines)):
    Close.append(float(klines[Number][4])) # На 4 позиции находится Close свечки

  LastKlines = klines[len(klines)-1][6]
  if LastKlines > NormalToTimestamp(TimeNow()):
    Close.pop(len(Close)-1)

  LocalTimeSpent_start = TimeSpent('Затраченное время на запись всех Close в list: ',LocalTimeSpent_start)

  data = {'Close': Close}
  data = pd.DataFrame(data)
  data = data.append({'Close':69,},ignore_index=True)

  LocalTimeSpent_start = TimeSpent('Затраченное время на преоброзование из list в DataFrame: ',LocalTimeSpent_start)
  TotalTimeSpent_end = TimeSpent('Полное затраченное время на загрузку и формирование бд: ',TotalTimeSpent_start)

  return TotalTimeSpent_start,data

def CloseChngCreate(data):
  _TimestampNow_ = TimeStampNow()
  TotalTimeSpent_start = _TimestampNow_
  LocalTimeSpent_start = _TimestampNow_
  for i in range(1, step_in_the_last + 1):
    indicator_name = 'Close_chng_%d' % (i)
    data[indicator_name] = data['Close'].pct_change(i) # относительная доходность единицах
  data = data.dropna() # удаляем строки с NaN

  LocalTimeSpent_start = TimeSpent('Создание разметки направления тренда: ',LocalTimeSpent_start)

  # Добавил разметку направления тренда
  for i in range(1, step_in_the_last + 1):
    indicator_name = 'Close_chng_%d' % (i)
    indicator_name_new = 'Y_%d' % (i)
    data[indicator_name_new] = 0
    data.loc[data[indicator_name] > 0 , indicator_name_new] = 1

  LocalTimeSpent_start = TimeSpent('Добавление разметки в DataFrame: ',LocalTimeSpent_start)

  for i in range(1, step_in_the_last + 1):
    indicator_name = 'Close_chng_%d' % (i)
    del data[indicator_name]
  indicator_name = 'Y_%d' % (step_in_the_last)
  data_Y = data[indicator_name]
  categorical_labels = to_categorical(data_Y, num_classes = 2)
  print(data)
  print(data_Y)
  print(categorical_labels)
  print(len(categorical_labels))

  LocalTimeSpent_start = TimeSpent('Редактирование разметки: ',LocalTimeSpent_start)
  TotalTimeSpent_start = TimeSpent('Полное затраченное время на загрузку и формирование финального DataFrame: ',TotalTimeSpent_start)

  return data,categorical_labels

def split_sequence(sequence, Y, n_steps, forvard_lag):
  X, y = list(), list()
  for i in range(len(sequence)):
    end_ix = i + n_steps # находим конечный индекс строки
    if end_ix + (forvard_lag-1) > len(sequence)-1: # проверяем, чтобы не выйти за пределы массива
      break
    seq_x, seq_y = sequence[i:end_ix], Y[end_ix + (forvard_lag - 1)]
    X.append(seq_x)
    y.append(seq_y) # тк предсказываем только Close
  return np.array(X), np.array(y)

def CreatingTensor(data,categorical_labels,forvard_lag=forvard_lag,n_steps=n_steps):
  _TimestampNow_ = TimeStampNow()
  TotalTimeSpent_start = _TimestampNow_
  LocalTimeSpent_start = _TimestampNow_
  data = data.values
  xTest = data
  yTest = categorical_labels
  xScaler = StandardScaler()
  xScaler.fit(xTest)
  xTest = xScaler.transform(xTest)
  LocalTimeSpent_start = TimeSpent('Скалярное перемножение для формирования тензора: ',LocalTimeSpent_start)
  for i in [forvard_lag]:
    # Формируем выборки для обучения нейросети
    xTest, yTest = split_sequence(xTest, yTest, n_steps, i)
  LocalTimeSpent_start = TimeSpent('Формируем выборки для обучения нейросети: ',LocalTimeSpent_start)
  TotalTimeSpent_start = TimeSpent('Полное затраченное время на создание финальной выборки для проверки: ',TotalTimeSpent_start)
  return TotalTimeSpent_start,xTest,yTest,data

def BUYtoSELL(_side_,fixed_btc,bot=bot):
  print(f'{_side_}/Time:', TimestampToNormal(bot.futuresCreateOrder(
    symbol=_SYMBOL_,
    side=_side_,
    type='MARKET',
    quantity=fixed_btc,
  )['updateTime']))

def BUY_a(fixed_btc=fixed_btc):
  global BUY_A
  while True:
    time.sleep(0.01)
    if BUY_A:
      BUY_A = False
      BUYtoSELL('BUY',fixed_btc)

def BUY_b(fixed_btc=fixed_btc):
  global BUY_B
  while True:
    time.sleep(0.01)
    if BUY_B:
      BUY_B = False
      BUYtoSELL('BUY',fixed_btc)

def SELL_a(fixed_btc=fixed_btc):
  global SELL_A
  while True:
    time.sleep(0.01)
    if SELL_A:
      SELL_A = False
      BUYtoSELL('SELL',fixed_btc)

def SELL_b(fixed_btc=fixed_btc):
  global SELL_B
  while True:
    time.sleep(0.01)
    if SELL_B:
      SELL_B = False
      BUYtoSELL('SELL',fixed_btc)

def CalculateProfit(data,TtF,_prediction_,Profit,Real,number,fix=fix,Commission_sell=Commission_sell,Commission_buy=Commission_buy):
  profit_percentage = abs(data[n_steps+number-4][0] - data[n_steps+number][0]) / (data[n_steps+number][0])
  print(data[n_steps+number-step_in_the_last][0], data[n_steps+number][0])
  fix_precentage = fix * profit_percentage
  if _prediction_ == 0:
    Commission = Commission_sell*fix + Commission_buy*(fix-fix_precentage)
    Profit_new = Profit + fix_precentage*TtF - Commission
  else:
    Commission = Commission_sell*(fix+fix_precentage) + Commission_buy*fix
    Profit_new = Profit + fix_precentage*TtF - Commission
  print("")
  print('Prediction:', _prediction_, 'Real:', np.argmax(Real), 'Заработал:', f'{round(Profit_new-Profit,6)}$', 'Заработал всего:', f'{round(Profit_new,6)}$' ,'Комиссия:', f'{round(Commission,6)}$')
  print("")
  return Profit_new

def GetPrediction(x_test,number,model=model):
  x = x_test[number]
  x = np.expand_dims(x, axis=0)
  prediction = model.predict(x) # Распознаём наш пример
  prediction = np.argmax(prediction) # Получаем индекс самого большого элемента (это итоговая цифра)
  return prediction

def PrintPrediction(x_test,y_test_org,data,Profit,number):   # Выбираю пример
  prediction = GetPrediction(x_test,number+1)
  if prediction == np.argmax(y_test_org[number+1]):
    TtF = 1
    _TtF_ = 1
  else:
    TtF = -1
    _TtF_ = 0

  Profit = CalculateProfit(data,TtF,prediction,Profit,y_test_org[number+1],number)
  return Profit,_TtF_

def CheckingResponses(xTest,yTest,data,Old_Prediction,Profit):
  global BUY_A
  global BUY_B
  global SELL_A
  global SELL_B
  global last
  global first
  _TimestampNow_ = TimeStampNow()
  TotalTimeSpent_start = _TimestampNow_
  LocalTimeSpent_start = _TimestampNow_
  number = len(xTest)-1
  prediction = GetPrediction(xTest,number)
  print(prediction)
  if first:
    if Old_Prediction == 0:
      pass
      BUY_A = True
      #КУПИТЬ
      print("Купить")
    else:
      pass
      SELL_A = True
      #ПРОДАТЬ
      print("Продать")
  Old_Prediction = prediction
  if last:
    if Old_Prediction == 0:
      #ПРОДАТЬ
      SELL_B = True
      first = True
      print("Продать")
      pass
    else:
      #КУПИТЬ
      BUY_B = True
      print("Купить")
      first = True
      pass
  LocalTimeSpent_start = TimeSpent('Затраченное время на транзакции: ',LocalTimeSpent_start)
  number = len(xTest)-2
  Profit,_TtF_  = PrintPrediction(xTest,yTest,data,Profit,number)
  LocalTimeSpent_start = TimeSpent('Высчитывание профита: ',LocalTimeSpent_start)
  TotalTimeSpent_start = TimeSpent('Полное затраченное время на транзакции и высчитывания профита: ',TotalTimeSpent_start)
  return Profit,Old_Prediction,_TtF_

def CheckingResponses_test(xTest,yTest,data,Old_Prediction,Profit,first=first):
  global BUY_A
  global BUY_B
  global SELL_A
  global SELL_B
  global last
  _TimestampNow_ = TimeStampNow()
  TotalTimeSpent_start = _TimestampNow_
  LocalTimeSpent_start = _TimestampNow_
  number = 1
  prediction = GetPrediction(xTest,number)
  if first:
    if Old_Prediction == 0:
      pass
      BUY_A = True
      #КУПИТЬ
    else:
      pass
      SELL_A = True
      #ПРОДАТЬ
  Old_Prediction = prediction
  if last:
    if Old_Prediction == 0:
      #ПРОДАТЬ
      SELL_B = True
      first = True
      pass
    else:
      #КУПИТЬ
      BUY_B = True
      first = True
      pass
  LocalTimeSpent_start = TimeSpent('Затраченное время на транзакции: ',LocalTimeSpent_start)
  _ANSWER_ = []
  Profit_list = []
  for number in range(0,len(xTest)-forvard_lag-1):
    Profit,_TtF_  = PrintPrediction(xTest,yTest,data,Profit,number-1)
    LocalTimeSpent_start = TimeSpent('Высчитывание профита: ',LocalTimeSpent_start)
    _ANSWER_.append(_TtF_)
    Profit_list.append(Profit)
  TotalTimeSpent_start = TimeSpent('Полное затраченное время на транзакции и высчитывания профита: ',TotalTimeSpent_start)
  return Profit_list,Profit,Old_Prediction,_ANSWER_

list_def = [BUY_a,BUY_b,SELL_a,SELL_b]

def create_threading(list_def):
  list_threading = []
  for row in list_def:
    list_threading.append(Thread(target=row))
    print("Создан поток:", row)
  return list_threading

def start_threading(list_threading):
  for row in list_threading:
    row.start()
    print("Запущен поток:", row)

def LoadTrainData(_SYMBOL_=_SYMBOL_):
  data = pd.read_csv(f'/content/drive/My Drive/{_SYMBOL_}.csv')
  head_tail_slice = list(range(data.shape[0]-analyze_period,data.shape[0]-1))
  data = data.iloc[head_tail_slice]
  print(data)
  return data

def LoadTrainDataV2(_SYMBOL_=_SYMBOL_):
  data = pd.read_csv(f'/content/drive/My Drive/{_SYMBOL_}.csv')
  del data['Timestamp']
  head_tail_slice = list(range(data.shape[0]-analyze_period,data.shape[0]-1))
  data = data.iloc[head_tail_slice]
  print(data)
  return data

def SampleFormation(data,step_in_the_last,n_steps,n_train=n_train,forvard_lag=forvard_lag):
  #df_data = data.drop(data.columns[[1, 2, 3, 4, 6]], axis='columns') # Оставляем дату и цену, чтобы потом добавить в outputs
  del data['Date'], data['Open'], data['Low'], data['High']
  # считаем относительную доходность единицах на step_in_the_last шагов назад
  for i in range(1, step_in_the_last + 1):
    indicator_name = 'Close_chng_%d' % (i)
    data[indicator_name] = data['Close'].pct_change(i) # относительная доходность единицах
  data = data.dropna() # удаляем строки с NaN
  # Добавил разметку направления тренда
  for i in range(1, step_in_the_last + 1):
    indicator_name = 'Close_chng_%d' % (i)
    indicator_name_new = 'Y_%d' % (i)
    data[indicator_name_new] = 0
    data.loc[data[indicator_name] > 0 , indicator_name_new] = 1
  for i in range(1, step_in_the_last + 1):
    indicator_name = 'Close_chng_%d' % (i)
    del data[indicator_name]
  #del data['Open'], data['High'], data['Low']
  indicator_name = 'Y_%d' % (step_in_the_last) #step_in_the_last forvard_lag
  data_Y = data[indicator_name]
  categorical_labels = to_categorical(data_Y, num_classes = 2)
  data = data.values
  xTrain = data[:-n_train-n_steps]
  xTest = data[-n_train:]
  yTrain = categorical_labels[:-n_train-n_steps]
  yTest = categorical_labels[-n_train:]
  xScaler = StandardScaler()
  xScaler.fit(xTrain)
  xTrain = xScaler.transform(xTrain)
  xTest = xScaler.transform(xTest)
  for i in [forvard_lag]:
    # Формируем выборки для обучения нейросети
    xTrain, yTrain = split_sequence(xTrain, yTrain, n_steps, i)
  for i in [forvard_lag]:
    # Формируем выборки для обучения нейросети
    xTest, yTest = split_sequence(xTest, yTest, n_steps, i)
  print("")
  print(yTrain)
  print("")
  print(yTest)
  print("")
  return xTrain,yTrain,xTest,yTest,data

def CreatingTrainingANetwork(xTrain,yTrain,xTest,yTest):
  drop = 0.4
  input = Input(shape=(xTrain.shape[1], xTrain.shape[2]))
  x = Flatten()(input)
  x = RepeatVector(4)(x)
  x = Conv1D(n_steps*2, 5, padding='same', activation='relu')(x)
  x = MaxPooling1D(pool_size=2)(x)
  x = Flatten()(x)
  x = RepeatVector(4)(x)
  x = Conv1D(n_steps, 5, padding='same', activation='relu')(x)
  x = MaxPooling1D(pool_size=2)(x)
  x = Flatten()(x)
  x = Dense(n_steps*150, activation='relu')(x)
  x = Dropout(drop)(x)
  x = Dense(yTrain.shape[1], activation='sigmoid')(x)
  model = Model(input, x)
  #xTrain , yTrain = X, y # если не используется TimeSeries
  model.compile(optimizer=Adam(learning_rate = 5e-04),
                loss='binary_crossentropy',
                metrics=['accuracy'])
  # fit model
  history = model.fit(xTrain, yTrain, epochs=5, verbose = 1,
                      validation_data=(xTest, yTest))
  return model

def CheckingTheNetwork(xTest,yTest,data,model,Variations,n_steps,n_train=n_train):
  data = data[-n_train:]
  # От себя добавил цикл для распознавания всех тестовых вариантов и вывода класса
  Test = n_train - n_steps
  # От себя добавил цикл для распознавания всех тестовых вариантов и вывода класса
  x_test = xTest[-Test:]
  y_test_org = yTest[-Test:]
  true_answer = 0
  for number in range(len(x_test)):   # Выбираю пример
    x = x_test[number]
    x = np.expand_dims(x, axis=0)
    prediction = model.predict(x) # Распознаём наш пример
    prediction = np.argmax(prediction) # Получаем индекс самого большого элемента (это итоговая цифра)
    if prediction == np.argmax(y_test_org[number]):
      true_answer+=1
  Variations.append({"True Answer":(true_answer/Test)*100, "Model":model, "Step_in_the_last":step_in_the_last,"N_steps":n_steps})
  return Variations
```
